### Box plot & strip plot

```python
plt.figure(figsize=(25,20))
plt.subplot(2,2,1)
sns.boxplot(x='Species', y='PetalWidthCm' , data=iris ) 
sns.stripplot(x='Species', y='PetalWidthCm' , data=iris, color='black')
plt.subplot(2,2,2)
sns.boxplot(x='Species', y='PetalLengthCm' , data=iris)
sns.stripplot(x='Species', y='PetalLengthCm' , data=iris,color='black')
plt.subplot(2,2,3)
sns.boxplot(x='Species', y='SepalWidthCm' , data=iris)
sns.stripplot(x='Species', y='SepalWidthCm' , data=iris ,color='black')
plt.subplot(2,2,4)
sns.boxplot(x='Species', y='SepalLengthCm' , data=iris)
sns.stripplot(x='Species', y='SepalLengthCm' , data=iris, color='black')
plt.xticks(rotation=45)
plt.xlabel('Species_Data')
plt.ylabel('Count')
fig=plt.gcf()
fig.set_size_inches(20,15)
```

#### Violinplot - It is used to visualize the distribution of data and its probability distribution .This chart is a combination of box plot and density plot that is rotated and placed each side , to the distrubution shape of the data. Thick black bar in the centre represents the interquartile range , the thin black line extended from it represents the 95% confidence intervals , and the white is median box. Plots are limited in their display of the data , as their visual simplicity tends to hide significant detials about how values in the data are distributed.

```python
plt.figure(figsize=(25,20))
plt.subplot(2,2,1)
sns.violinplot(x='Species', y='PetalWidthCm' , data=iris)
#sns.stripplot(x='Species', y='PetalWidthCm' , data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='Species', y='PetalLengthCm' , data=iris)
#sns.stripplot(x='Species', y='PetalLengthCm' , data=iris,edgecolor = 'red')
plt.subplot(2,2,3)
sns.violinplot(x='Species', y='SepalWidthCm' , data=iris)
#sns.stripplot(x='Species', y='SepalWidthCm' , data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='Species', y='SepalLengthCm' , data=iris)
#sns.stripplot(x='Species', y='SepalLengthCm' , data=iris)
plt.xticks(rotation=45)
plt.xlabel('Species_Data')
plt.ylabel('Count')
fig=plt.gcf()
fig.set_size_inches(20,15)
```

#### Pairplot : A 'pair plot' is also known as a scatterplot , in which one variable int he same data row is matched with another variable's value, like this. Pair plots are just elabrations on this , showing all variables paired with the another variable.

#### Heatmap : Heatmap is used to find the out the correlaion between differant features in the dataset.high positive or negative value shows that the feature have high co relation. this helps us to select the parameters for the machine learning.

```python

f , ax = plt.subplots(figsize=(12, 9), )
sns.heatmap(iris.corr() , cmap = 'YlGnBu',linewidths=.5 , vmax=1,vmin=-1, square=True) 
plt.xticks(rotation=90)
plt.yticks(rotation=360)
f.savefig("output.png")
fig=plt.gcf()
fig.set_size_inches(10,5)
```

#### Swarm plot It looks a bit like a friendly swarm of bees buzzing about their hive. More importantly, each data point is clearly visible and no data are obscured by overplotting.A beeswarm plot improves upon the random jittering approach to move data points the minimum distance away from one another to avoid overlays. The result is a plot where you can see each distinct data point, like shown in below plot

```python
plt.figure(figsize=(10,8))
sns.swarmplot(x='Species', y='PetalWidthCm' , data=iris)
plt.xticks(rotation=45)
plt.xlabel('Species_Data')
plt.ylabel('Relation')


fig=plt.gcf()
fig.set_size_inches(10,8)
```

### Boxplot vs Swarmplot

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Species', y='SepalLengthCm' , data=iris )
sns.swarmplot(x='Species', y='SepalLengthCm' , data=iris, color='black')
plt.xticks(rotation=45)
plt.xlabel('Species_Data')
plt.ylabel('Relation')
fig=plt.gcf()
fig.set_size_inches(10,8)
```

### violinplot vs Swarmplot

```python
sns.violinplot(x='Species', y='PetalLengthCm' , data=iris)
sns.swarmplot(x='Species', y='PetalLengthCm' , data=iris, color='black')
plt.xticks(rotation=45)
plt.xlabel('Species_Data')
plt.ylabel('Relation')
fig=plt.gcf()
fig.set_size_inches(10,8)
```

### Species based classification using set

```python
sns.set(style="darkgrid")
sc=iris[iris.Species=='Iris-setosa'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='red',label='Setosa')
iris[iris.Species=='Iris-versicolor'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='green',label='Versicolor',ax=sc)
iris[iris.Species=='Iris-virginica'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='orange', label='virginica', ax=sc)
sc.set_xlabel('Sepal Length in cm')
sc.set_ylabel('Sepal Width in cm')
sc.set_title('Sepal Length Vs Sepal Width')
sc=plt.gcf()
sc.set_size_inches(10,6)
```

#### LmPlot - Is like a linear regresion with scatter ploting, In other words, The lineplot (lmplot) is one of the most basic plots. It shows a line on a 2 dimensional plane. You can plot it with seaborn or matlotlib depending on your preference.The lmplot plot shows the line along with datapoints on the 2d space. By specifying x and y you can set the horizontal and vertical labels repsectively.

```python
fig=sns.lmplot(x="PetalLengthCm", y="PetalWidthCm",data=iris)
```

#### FacetGrid - FacetGrid object takes a dataframe as input and the names of the variables that will form the row, column, or hue dimensions of the grid.The variables should be categorical and the data at each level of the variable will be used for a facet along that axis.

```python
sns.FacetGrid(iris, hue="Species", size=6) \
   .map(sns.distplot, "PetalLengthCm") \
   .add_legend()
plt.ioff()
```

### Factor plot - Factor plot is a great method to visualize different categorical values, but it is tricky to add the title and change the size, or manipulate x and y axes.

```python
sns.factorplot('Species','SepalLengthCm',data=iris)
plt.ioff()
plt.show()
```

#### Boxenplot :  The Boxen plot is very similar to box plot, except for the fact that it plots different quartile values. By plotting different quartile values, we are able to understand the shape of the distribution particularly in the head end and tail end. 

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxenplot(x='Species',y='SepalLengthCm',data=iris)
```
