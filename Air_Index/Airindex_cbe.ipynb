```python
uni_chn_station=list(df_chn.STATION.unique())
uni_chn_station
```

```python
tmp_df = []

for station in uni_chn_station:
    tmp_chn = df_chn[df_chn['STATION'] == station ].groupby(["Remarks"])[['AQI']].count().sort_values('AQI',ascending=False).reset_index()
    tmp_chn['station_name']= station
    tmp_df.append(tmp_chn)
```

```python
merge_df= pd.concat(tmp_df).reset_index()
merge_df
```

```python
uni_chn_typ=list(df_chn.Type.unique())
uni_chn_typ
```

```python
df_type = []

for typ in uni_chn_typ:
    typ_chn = df_chn[df_chn['Type'] == typ ].groupby(["Remarks"])[['AQI']].count().sort_values('AQI',ascending=False).reset_index()
    typ_chn['typ_name']= typ
    df_type.append(typ_chn)
    
merge_typ = pd.concat(df_type)
merge_typ[merge_typ['Remarks']== 'Satisfactory'].sort_values('AQI',ascending=False).reset_index()
```
